```python
import matplotlib.pyplot as plt
import numpy as np 
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    '/Data Science Softwares/Plant Detection/archive_3/test',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
cnn=tf.keras.models.load_model("/Data Science Softwares/Plant Detection/planttrinedmodel.h5")
```

```python
import cv2
imagepath='/Data Science Softwares/Plant Detection/archive_3/test/onion/Image_4.jpg'
img=cv2.imread(imagepath)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image=tf.keras.preprocessing.image.load_img(imagepath,target_size=(64,64))
input_array=tf.keras.preprocessing.image.img_to_array(image)
input_array=np.array([input_array])
prediction=cnn.predict(input_array)
```

```python
print(prediction)
```

```python
result_index = np.argmax(prediction) #Return index of max element
print(result_index)
```

```python
# Displaying the image 
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#Single image Prediction
print("It's a {}".format(test_set.class_names[result_index]))
```
